```python
cantinuous_columns = ["age","hypertension","heart_disease","bmi","HbA1c_level","blood_glucose_level","diabetes"]
categorical_columns = ["gender","smoking_history"]
```

```python
from scipy import stats
```

```python
def two_sample_t_test(d1,d2):
  t = 0
  f = 0
  for i in range(31):
    sample1 = d1.sample(frac=0.04)
    sample2 = d2.sample(frac =0.04)
    t_test,p_value = stats.ttest_ind(sample1,sample2)
    if p_value < 0.05:
      f += 1
    else:
      t += 1
    if f > t:
      return True
    else:
      return False
```

```python
two_sample_t_test(df['age'],df['diabetes'])
```

```python
two_sample_t_test(df['age'],df['hypertension'])
```

```python
two_sample_t_test(df['age'],df['heart_disease'])
```

```python
two_sample_t_test(df['age'],df['bmi'])
```

```python
two_sample_t_test(df['age'],df['HbA1c_level'])
```

```python
two_sample_t_test(df['age'],df['blood_glucose_level'])
```

```python
def chi_square_test_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.05 else False
```

```python
chi_square_test_cat_vs_cat(df['gender'],df['smoking_history'])
```

```python
def annova_test(d1,d2):
  group = []
  for i in d1.unique():
    group.append(d2[d1 == i])
  return True if stats.f_oneway(*group)[1] < 0.05 else False
```

```python
annova_test(df['gender'],df['age'])
```

```python
annova_test(df['gender'],df['hypertension'])
```

```python
final = {}
for i in df.columns:
  final[i] = {}
  for j in df.columns:
    if (i in cantinuous_columns) and (j in cantinuous_columns):
      final[i][j] = two_sample_t_test(df[i],df[j])
    elif (i in categorical_columns) and (j in cantinuous_columns):
      final[i][j] = annova_test(df[i],df[j])
    elif (i in cantinuous_columns) and (j in categorical_columns):
      final[i][j] = annova_test(df[i],df[j])
    elif (i in categorical_columns) and (j in categorical_columns):
      final[i][j] = chi_square_test_cat_vs_cat(df[i],df[j])
    if final:
      final[i][j]=1
    else:
      final[i][j]=0
```
